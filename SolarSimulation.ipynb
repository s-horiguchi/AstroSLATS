```python
dn = 3*30+20
long = 0 #diff from JST
lat = 35

theta0 = 2*np.pi / 365 * (dn-1)
delta = 0.006918-0.399912*np.cos(theta0)+0.070257*np.sin(theta0)-0.006758*np.cos(2*theta0)+ \
    0.000907*np.sin(2*theta0)-0.002697*np.cos(3*theta0)+0.001480*np.sin(3*theta0)
Eq = 0.000075+0.001868*np.cos(theta0)-0.032077*np.sin(theta0)-0.014615*np.cos(2*theta0)-0.040849*np.sin(2*theta0)
t = np.linspace(0,24,100)
h=(t-12)*np.pi/12.+0+Eq
alpha = np.arcsin(np.sin(lat)*np.sin(delta) + np.cos(lat)*np.cos(delta)*np.cos(h))
psi = np.arctan2(np.cos(lat)*np.cos(delta)*np.sin(h), np.sin(lat)*np.sin(alpha)-np.sin(delta))
```

```python
plt.plot(t, -alpha/np.pi*180)
```
